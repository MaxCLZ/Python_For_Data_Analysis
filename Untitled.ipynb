## Data Cleaning
### Recherche des valeurs manquantes

On va dans un premier temps checker les colonnes où il y a beaucoup trop de valeurs manquantes

```python
line_count=len(dataf["number"])
count=0
for i in dataf.columns:
    count=0
    for y in range(len(dataf[i])):
        if(dataf[i][y]=='?'):
            count+=1
    if(count/line_count*100 > 50): 
        print(Back.RED + "WARNING column :",i," percent of missing data superior to 50 % , percentage :", count/line_count*100)
        print(Style.RESET_ALL)
    elif(count/line_count*100 > 0):
        print(Back.BLUE + "WARNING column :",i," there is some missing data, percentage :", count/line_count*100)
        print(Style.RESET_ALL)
    else:
        print("Column :",i,"there is no missing data on this column")

```

On remarque donc que les colonnes suivantes sont à plus de 50% en pourcentages de données manquantes:
- cmdb_ci
- problem_id
- rfc
- vendor
- caused_by

On va donc les supprimer du dataset

On remarque aussi les colonnes où il manque des données:
- caller_id
- opened_by
- sys_created_by
- sys_created_at
- location
- category 
- subcategory
- u_symptom
- assignment_group
- assigned_to  
- closed_code  
- resolved_by
- resolved_at

On va donc devoir remplir les trous de données

```python
print("Nombre de colonnes du dataset avant drop",dataf.shape[1]) 

dataf.drop(columns=["cmdb_ci", "problem_id", "rfc", "vendor", "caused_by"], inplace=True)

print("Nombre de colonnes du  dataset après drop",dataf.shape[1]) 
```

```python
new_dataf= dataf.copy()
new_dataf.drop(columns=["cmdb_ci", "problem_id", "rfc", "vendor", "caused_by"], inplace=True)

print("Nombre de colonnes du dataset de base",dataf.shape[1]) 
print("Nombre de colonnes du  nouveau dataset de base",new_dataf.shape[1]) 
```

On va donc maintenant s'occuper des colonnes où il manque des données
On va remplir les trous par la valeur NaN pour les colonnes qui ne sont pas des dates 
On définit une fonction pour faciliter cette étape


```python
new_dataf['caller_id']=new_dataf['caller_id'].apply(lambda x: np.nan if x == '?'else x)
new_dataf['opened_by']=new_dataf['opened_by'].apply(lambda x: np.nan if x == '?'else x)
new_dataf['sys_created_by']=new_dataf['sys_created_by'].apply(lambda x: np.nan if x == '?'else x)
new_dataf['sys_created_at']=new_dataf['sys_created_at'].apply(lambda x: np.nan if x == '?'else x)
new_dataf['location']=new_dataf['location'].apply(lambda x: np.nan if x == '?'else x)
new_dataf['category']=new_dataf['category'].apply(lambda x: np.nan if x == '?'else x)
new_dataf['subcategory']=new_dataf['subcategory'].apply(lambda x: np.nan if x == '?'else x)
new_dataf['u_symptom']=new_dataf['u_symptom'].apply(lambda x: np.nan if x == '?'else x)
new_dataf['assignment_group']=new_dataf['assignment_group'].apply(lambda x: np.nan if x == '?'else x)
new_dataf['assigned_to']=new_dataf['assigned_to'].apply(lambda x: np.nan if x == '?'else x)
new_dataf['closed_code']=new_dataf['closed_code'].apply(lambda x: np.nan if x == '?'else x)
new_dataf['resolved_by']=new_dataf['resolved_by'].apply(lambda x: np.nan if x == '?'else x)
new_dataf['resolved_at']=new_dataf['resolved_at'].apply(lambda x: np.nan if x == '?'else x)
```

```python
line_count=len(new_dataf["number"])
count=0
for i in new_dataf.columns:
    count=0
    for y in range(len(new_dataf[i])):
        if(new_dataf[i][y]=='?'):
            count+=1
    if(count/line_count*100 > 0):
        print(Back.RED + "WARNING column :",i," there is some missing data, percentage :", count/line_count*100)
        print(Style.RESET_ALL)
    else:
        print(Back.GREEN +"Column :",i,"there is no missing data on this column")
        print(Style.RESET_ALL)

```

### Le but de ce dataset est de prédire la durée de résolution d'un incident
On va donc devoir adapter les données de ce dataset

```python
print(type(new_dataf['opened_at']))
print(type(new_dataf['resolved_at']))
print(type(new_dataf['closed_at']))
print(type(new_dataf['sys_updated_at']))
print(type(new_dataf['sys_created_at']))
```

On va convertir ces colonnes là pour avoir des variable de type DateTime pour pouvoir les manipuler plus aisément 

```python
new_dataf['opened_at']=pd.to_datetime(new_dataf['opened_at'], dayfirst=True)
new_dataf['resolved_at']=pd.to_datetime(new_dataf['resolved_at'], dayfirst=True)
new_dataf['closed_at']=pd.to_datetime(new_dataf['closed_at'], dayfirst=True)
new_dataf['sys_updated_at']=pd.to_datetime(new_dataf['sys_updated_at'], dayfirst=True)
new_dataf['sys_updated_at']=pd.to_datetime(new_dataf['sys_updated_at'], dayfirst=True)
```
